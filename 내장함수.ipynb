```python
#all(x)
print(all([1,2,3])) #요소가 모두 참이면 True
print(all([1,2,3,0])) #거짓 하나라도 있으면 False
```

```python
#any(x)
print(any([1,2,3])) #요소 중 하나라도 참이면 True
print(any(["",0])) #거짓
```

```python
#str(object) : 객체를 문자열로 반환
str(3)
```

```python
#chr(i) : 유니코드 숫자 값을 입력받아 해당하는 문자를 반환하는 함수
chr(97), chr(44032)#아스키코드를 문자로 변환
```

```python
#ord(c) : 문자의 유니코드 숫자값 반환
ord('a'), ord('가')
```

```python
#dir(x) : 객체가 지닌 변수나 함수를 보여준다 (사용할 함수 찾을 때 유용할듯 함)
print(dir([1,2,3]))
print(dir({'1':'a'}))
```

```python
#divmod(a,b) : a 나누기 b의 몫과 나머지를 튜플로 반환함
divmod(7,3)
```

```python
#enumerate: 열거하다. 인덱스 값을 포함하는 enumrate객체를 반환함
for i, name in enumerate(['a','b','c']):
    print(i,name)
```

```python
#eval(expression) : 문자열 표현식을 입력받아 실행한 결과값을 반환하는 함수다.
print(eval('1+2'))
print(eval("'hi'+'a'"))
print(eval('divmod(4,3)'))
```

```python
#filter(함수, 반복_가능한_데이터)
def positive(x):
  return x > 0
print(list(filter(positive,[1,-3,2,0,-5,6])))

list(filter(lambda x: x>0,[1,-3,2,0,-5,6])) #lambda로 더 간결하게
```

```python
#int(x, radix) : radix진수를 10진수 정수로 반환 (radix생략 가능)
int("153"), int(3.56), int('111',2)
```

```python
#hex(x) : 정수 입력 -> 16진수 문자열 반환
hex(10)
```

```python
#oct(x) : 정수 입력 -> 8진수 문자열 반환
oct(34)
```

```python
#id(object) : 객체 입력 -> 주소값 반환
a = 3
id(a), id(3)
```

```python
#isinstance(object, class) 객체가 해당 클래스의 인스턴스인지 판단
class Person: pass
a = Person()
isinstance(a, Person)
```

```python
#len(s)
len('python')
```

```python
#list(iterable)
list('python')
```

```python
#map(f,iterable) : 함수(f)와 반복가능데이터 입력 -> 각 요소에 f를 적용하여 반환
def two_mul(x):
  return x*2
print(list(map(two_mul, [1,2,3,4])))
#lambda사용
list(map(lambda x: x*2, [1,2,3,4]))
```

```python
#max(iterable) : 최댓값 반환
max([1,2,60,3,4]), max('python')
```

```python
#min(iterable) : 최솟값 반환
min([1,2,60,3,4]), min('python')
```

```python
#abs(x)
print(abs(-3)) #절댓값 반환
```

```python
#pow(x,y) : x**y
pow(2,4), pow(3,3)
```

```python
#zip(*iterable) 동일한 개수로 이루어진 데이터들을 묶어서 반환
list(zip([1,2,3],[4,5,6]))
```
